```python
import numpy as np
import pandas as pd
import seaborn as sn
```

```python
dataset = pd.read_csv('Asteroid_Updated.csv')
```

```python
dataset = dataset.drop(["extent","rot_per","GM","BV","UB","IR","spec_B","spec_T","G"],axis=1)
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset["e"].fillna(dataset["e"].mean(),inplace=True)
dataset["i"].fillna(dataset["i"].mean(),inplace=True)
dataset["om"].fillna(dataset["om"].mean(),inplace=True)
dataset["w"].fillna(dataset["w"].mean(),inplace=True)
dataset["q"].fillna(dataset["q"].mean(),inplace=True)
dataset["ad"].fillna(dataset["ad"].mean(),inplace=True)
dataset["per_y"].fillna(dataset["per_y"].mean(),inplace=True)
dataset["data_arc"].fillna(dataset["data_arc"].mean(),inplace=True)
dataset["condition_code"].fillna(dataset["condition_code"].mean(),inplace=True)
dataset["n_obs_used"].fillna(dataset["n_obs_used"].mean(),inplace=True)
dataset["H"].fillna(dataset["H"].mean(),inplace=True)
dataset["moid"].fillna(dataset["moid"].mean(),inplace=True)
dataset["n"].fillna(dataset["n"].mean(),inplace=True)
dataset["per"].fillna(dataset["per"].mean(),inplace=True)
dataset["ma"].fillna(dataset["ma"].mean(),inplace=True)
```
